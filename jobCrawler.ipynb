```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import json

jobs = []

while True:

    try:
        #解析html
        soup = BeautifulSoup(driver.page_source, "html.parser")

        #列出工作
        for job in soup.find_all('a', class_=re.compile(r'jb-link jb-link-blue job-name')):

            print(job.text+":"+job['href'])
            jobs.append({'job_name': job.text, 'job_link': job['href']})

        # 尋找「下一頁」按鈕
        next_page = driver.find_element(By.XPATH, "//span[text()='下一頁']")

        # 點擊「下一頁」按鈕
        driver.execute_script("arguments[0].click();", next_page)

        # 等待頁面加載
        time.sleep(3)  # 根據需要調整等待時間

    except Exception as e:

        print("已經到達最後一頁或按鈕不可用:", e)

        break  # 如果找不到按鈕，跳出循環

with open('jobs.txt', 'w', encoding="utf-8") as convert_file:
     convert_file.write(json.dumps(jobs,ensure_ascii=False))
```

# Job Description Scraping

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from bs4 import BeautifulSoup
import requests
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.headless = True
```

```python
import json

# 讀取包含 JSON 格式的文本檔案
with open('jobs.txt', 'r', encoding='utf-8') as file:
    jobs = json.load(file)  # 使用 json.load 將檔案的內容轉換為 Python 字典或列表

# 打印讀取的 JSON 數據
print(jobs)
```

```python
import re

#創建WebDriver 實例：
driver = webdriver.Chrome(options=chrome_options)

#job_list = []

# 開始爬取並保存數據
with open("job_description.txt", "a", encoding="utf-8") as file:

  try:

      for job in jobs:

          job_data = {}

          try :
            #工作資訊
            url = 'https:'+job['job_link']
            driver.get(url)
            time.sleep(5)
            source = driver.page_source
            soup = BeautifulSoup(source, "html.parser")
            print(url)

          except Exception as e:
            print(e)
            print(url)
            print('工作連結出錯')

          try:
              pattern = r"(?<=job\/)[^?]+"
              result = re.search(pattern, url).group()

              job_data['job_no'] = result

          except Exception as e:
            print(e)

          try :
              job_data['job_company'] = soup.select_one('div.mb-3.bg-white').text.strip()

          except Exception as e:
              print("job_company")
              print(e)
          try :
              job_data['job_company_link'] =  soup.select_one('div.mb-3.bg-white a')['href']

          except Exception as e:
              print("job_company_link")
              print(e)

          try :
              job_data['job_name'] =  soup.select_one('div.bg-white.p-4 h1').text.strip()
          except Exception as e:

              print("job_name")
              print(e)
          try :
              salary_1 = soup.select_one('span.t3.mb-0.job-header__monthly-salary-remark.text-gray-deep-dark').text.strip()

          except Exception as e:
              salary_1 =""
              print("job_salary_1")
              print(e)

          try :
              salary_2 = soup.select_one('span.align-baseline').text.strip()

          except Exception as e:
              salary_2 =""
              print("job_salary_2")
              print(e)
          job_data['job_salary'] = salary_1 + salary_2


          try :
              salary_2 = soup.select_one('span.align-baseline').text.strip()

          except Exception as e:
              print("job_salary_2")
              print(e)

          try :
              job_data['job_description'] =soup.select_one('p.mb-5.r3.job-description__content.text-break').text.strip()

          except Exception as e:
              print("job_description")
              print(e)

          try :
              job_data['job_location'] = soup.select_one('div.job-address span').text.strip()

          except Exception as e:
              print("job_description")
              print(e)

          #try :
          #    job_data['job_tools'] =soup.select('div.job-requirement-table.row div.list-row.row.mb-2')[4].select('div.col p-0 list-row__data div')#.text.strip()

          #except Exception as e:
          #    print("job_tools")
          #    print(e)

          try :
              job_data['job_requirements'] = soup.select_one('p.m-0.r3.w-100').text.strip()

          except Exception as e:
              print("job_requirements")
              print(e)

          try :
              job_data['job_welfare'] =  soup.select_one('div.row.benefits-description div p').text.strip()

          except Exception as e:
              print("job_welfare")
              print(e)

          #公司資訊
          company_url = job_data['job_company_link']
          driver.get(company_url)
          time.sleep(5)
          source = driver.page_source
          soup = BeautifulSoup(source, "html.parser")

          try :
              job_data['company_intro'] = soup.select_one('div.col.serve-mobile__desc p').text.strip()

          except Exception as e:
              print("company_intro")
              print(e)

          try :
              job_data['company_product'] = soup.select_one('div.col.serve-mobile__desc p').text.strip()

          except Exception as e:
              print("company_product")
              print(e)
          try :
              job_data['company_location'] = soup.select_one('div.col.serve-mobile__desc p').text.strip()

          except Exception as e:
              print("company_location")
              print(e)

          #job_list.append(job_data)
          print(job_data)
          file.write(str(job_data) + "\n")

  except Exception as e:
          # 捕捉錯誤並繼續
          print(f"無法抓取，錯誤原因：{e}")


#with open('job_description.txt', 'w', encoding="utf-8") as convert_file:
     #convert_file.write(json.dumps(job_list,ensure_ascii=False))

```
